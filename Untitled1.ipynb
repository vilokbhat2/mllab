```python
from scikitplot.metrics import plot_roc_curve
from sklearn.metrics import roc_curve
```

```python
import matplotlib.pyplot as plt
```

```python
y_true = [1, 0, 1, 0]
y_pred = [0.1, 0.9, 0.87, 0.67]
```

```python
fpr, tpr, _ = roc_curve(y_true,  y_pred)
```

```python
plt.plot(fpr)
plt.plot(tpr)
plt.xlabel("fpr")
plt.ylabel("tpr")
plt.legend()
```
